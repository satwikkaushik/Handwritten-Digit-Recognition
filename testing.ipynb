```python
import os
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Load the model

```python
model = tf.keras.models.load_model('sheep_one.model')
```

# Load custom images and predict them

```python
def display_results(img, result):
    plt.imshow(img.squeeze(), cmap='gray')
    plt.axis('off')
    result_lbl = np.argmax(result)
    plt.title(f"Predicted: {result_lbl}")
    plt.show()
```

```python
img_count = 1
while os.path.isfile('samples/digit{}.png'.format(img_count)):
    try:
        img = cv2.imread('samples/digit{}.png'.format(img_count))[:,:,0]
        img = np.invert(np.array([img]))
        prediction = model.predict(img)
        display_results(img, prediction)
        img_count += 1
    except:
        print("Error in this img")
        img_count += 1
```
